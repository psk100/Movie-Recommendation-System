# Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(m_data['genres'])
```

```python
m_data.head()
```

```python
print(feature_vectors)
#based on each word it sepeatate in genras it create a seperate point
#in first we have five word for this word we create 5 diff float point
```

```python
import re
```

```python
for i in m_data['genres']:
    
    rp=re.sub('[^a-zA-Z]',' ',i)
    print(rp)
    break
```

```python
m_data['genres']=m_data['genres'].apply(lambda i:re.sub('[^a-zA-Z]',' ',i) )
```

```python
m_data
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity =cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
list_of_all_titles=m_data['movie_title'].tolist()
print(list_of_all_titles)
```

```python
import difflib 
def movies(movie_name):
    find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
    return find_close_match
```

```python
close=movies(input('Enter you Favaurite movie name:'))
#print(find_close_match[2])
close
```

```python
l=m_data.loc[m_data['movie_title']==close[0]].index.values[0]
```

```python
l              #finding the index no movie name 
```

```python
#getting a list of similar movies
```

```python
similarity_score=list(enumerate(similarity[l]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_sm=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_sm)
```

```python
len(sorted_sm)
```

# print the name of similar movies based on index 

```python
print('Movies suggested for you: \n')
for movie in sorted_sm:
    index=movie[0]
    title_index=m_data[m_data.index==index]['movie_title'].values[0]
    if index<=10:
        print(index,'-',title_index)
        
```
